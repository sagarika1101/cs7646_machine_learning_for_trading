```python
symbol = ['JPM']
start_date =  dt.datetime(2008,1,1)
end_date = 	 dt.datetime(2009,12,31)
market_dates = pd.date_range(start_date,end_date)
```

```python
start_val = 100000
impact = 0.005
commission = 9.95
```

```python
prices_JPM = get_data(symbol,market_dates)
prices_JPM = prices_JPM['JPM']
normed_prices = prices_JPM/prices_JPM.iloc[0]
normed_prices = normed_prices.to_frame()
```

```python
trades_df = pd.DataFrame(index=normed_prices.index)
```

```python
#initialzing trades_df values
trades_df['Symbol'] = 'JPM'
trades_df['Order'] = np.NaN
trades_df['Shares'] = 1000.0
```

```python
trades_df
```

```python
#calculate SMA and momentum
SMA = normed_prices.rolling(window=20,center=False).mean()
SMA = SMA.bfill()
momentum = (normed_prices - normed_prices.shift(20))/normed_prices.shift(20)
momentum = momentum.bfill()
```

```python
SMA
```

```python
momentum
```

```python
current_holdings = 0.0
for i in range(trades_df.shape[0]):
    #Check if SMA<0.95 or momentum<-0.5
    if (SMA.iloc[i,0] < 0.95 or momentum.iloc[i,0] < -0.5) and (current_holdings==0.0 or current_holdings==(-1000.0)):
        trades_df.iloc[i,1] = 'BUY'
        current_holdings+=1000.0
    #Check if SMA>1.05 or momentum>0.5
    elif (SMA.iloc[i,0] > 1.05 or momentum.iloc[i,0] > 0.5) and (current_holdings==0.0 or current_holdings==1000.0):
        trades_df.iloc[i,1] = 'SELL'
        current_holdings-=1000.0
    else:
        continue    
```

```python
#Remove days in which no trade is being made
trades_df.dropna(inplace=True)
#trades_df is the trades dataframe that we can pass to computer_portvals to compute protfolio values
```

```python
def compute_portvals(dfOrders, start_val = 100000, commission=0.0, impact=0.0, start_date = dt.datetime(2008,1,1), end_date = dt.datetime(2009,12,31)):  		   	  			  	 		  		  		    	 		 		   		 		  
    # this is the function the autograder will call to test your code  		   	  			  	 		  		  		    	 		 		   		 		  
    # NOTE: orders_file may be a string, or it may be a file object. Your  		   	  			  	 		  		  		    	 		 		   		 		  
    # code should work correctly with either input  		   	  			  	 		  		  		    	 		 		   		 		  
    # TODO: Your code here  	

    #read orders.csv into a dataframe and sort by date
    #dfOrders = pd.read_csv(orders_file, index_col='Date', parse_dates=True, na_values=['nan'])
    #dfOrders = dfOrders.sort_values('Date')

    #start_date =  dfOrders.index.values[0] 		
    #end_date = 	 dfOrders.index.values[-1]
    market_dates = pd.date_range(start_date,end_date)

    stocks = list(dfOrders.Symbol.unique())
    stockPrices = get_data(stocks,market_dates)

    stockPrices['portfolioVal'] = 0.0
    stockPrices['transactionCost'] = 0.0
    for i in stocks:
    	symShares = i + 'shares'
    	stockPrices[symShares] = 0.0
    stockPrices['totalCash'] = start_val
    stockPrices['isOrderMade'] = 0.0

    totalCash = start_val
    for index,row in dfOrders.iterrows():
    	stock = row['Symbol']
    	orderType = row["Order"]
    	shares = row['Shares']
    	symShares = stock + 'shares'
    	obj = stockPrices.loc[index]
    	transCost = commission + impact*shares*obj[stock]
    	obj['transactionCost'] = transCost
    	Cash = 0.0
    	if orderType=='BUY':
    		obj[symShares] += shares
    		Cash = totalCash - (shares*obj[stock]) - transCost
    		obj['totalCash'] = Cash
    		obj['isOrderMade'] += 1.0
    	else:
    		obj[symShares] -= shares
    		Cash = totalCash + (shares*obj[stock]) - transCost
    		obj['totalCash'] = Cash
    		obj['isOrderMade'] += 1.0
    	totalCash = Cash

    l = len(stocks)
    for i in range(1, stockPrices.shape[0]):
    	for j in range(len(stocks)):
    		stockPrices.iloc[i,(j+l+3)] += stockPrices.iloc[(i-1),(j+l+3)]
    		if stockPrices.iloc[i,-1]==0.0:
    			stockPrices.iloc[i,-2] = stockPrices.iloc[(i-1),-2]

    #update portfolioVal to represent portfolio value of each day
    for i in range(0,stockPrices.shape[0]):
    	portfolio = 0.0
    	#adjust according to number of stocks held that day
    	for j in range(len(stocks)):
    		portfolio += stockPrices.iloc[i,j+1]*stockPrices.iloc[i,(j+l+3)]
    	#add total cash
    	portfolio += stockPrices.iloc[i,-2]
    	#update portfolioVal column
    	#print(portfolio)
    	stockPrices.iloc[i,(l+1)] = portfolio

    portvals = stockPrices.iloc[:,(l+1)]
    portvals = portvals.to_frame()

    return portvals 		   
```

```python
portvals = compute_portvals(trades_df, 100000, 9.95, 0.005, start_date, end_date)
```
